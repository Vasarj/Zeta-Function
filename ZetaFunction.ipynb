# Riemann Zeta Function

```python
import matplotlib.pyplot as plt  # Pyplot graphing module
import numpy as np               # Module containing many useful (mostly numeric functions)
import datetime                  # Module for working with date and time

dat = datetime.datetime.now()    # Object containing current date and time
ymd = dat.strftime('%Y-%m-%d')   # String containing current date in format yyyy-mm-dd

f   = lambda x: np.sin(x)        # Riemann Zeta Function
pi  = np.pi

lw1 = 1.0                        # Linewidth (level 1)
lw2 = 0.75                       # Linewidth (level 2)
lw3 = 0.50                       # Linewidth (level 3)

N    = 129                       # Number of dsscrete points for plotting
xmin = -pi; xmax = pi;           # Limits on the x-axis
ymin = -1;  ymax =  1;           # Limits on the x-axis

x = np.linspace(xmin, xmax, N)   # Evenly spaced discrete on the x-axis

fig, ax1 = plt.subplots()                                # Create a figure containing a single axes
ax1.set_title('Riemann Zeta Function', color='blue')     # Set title
ax1.set_title('Vasily', loc='left')
ax1.set_title(ymd, loc='right')

clr = 'blue'                                                       # Set color for next plots
ln1 = ax1.plot(x, f(x), linewidth=lw1, color=clr, label='Zeta')    # Main line in ax1
ax1.plot([xmin, xmax], [0, 0], linewidth=lw3, color='black')
ax1.plot([0, 0], [ymin, ymax], linewidth=lw3, color='black')
ax1.set_aspect('equal', 'box')
#ax1.set_aspect('equal')

plt.show()
```
